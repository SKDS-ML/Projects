# Encoding

Ordinal Encoding to convert Categorical Values to Numerical Values

```python
from sklearn.preprocessing import OrdinalEncoder
enc=OrdinalEncoder()
```

```python
for i in df.columns:
    if df[i].dtypes=="object":
        df[i]=enc.fit_transform(df[i].values.reshape(-1,1))
```

# Removing Outliers using Zscore

```python
from scipy.stats import zscore

z=np.abs(zscore(df))

new_df=df[(z<3)]
```

Above command is to remove outliers using Zscore technique for the given dataset

```python
x=new_df.drop("Churn",axis=1)
y=new_df["Churn"]
```

# Importing  train_test_split

```python
from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score
```

# Identifying Best Random State

```python
for i in range(5):
    x_train, x_test, y_train, y_test = train_test_split (x,y,test_size=0.33, random_state=i)
    from sklearn.tree import DecisionTreeClassifier
    lm=DecisionTreeClassifier()
    lm.fit(x_train,y_train)
    pred=lm.predict(x_test)
    accu=accuracy_score(y_test,pred)
    print(accu*100, "at random_state", i)
```

```python
from sklearn.tree import DecisionTreeClassifier
maxAccu=0
maxRS=0
for i in range(1,5):
    x_train, x_test, y_train, y_test = train_test_split (x,y,test_size=0.30, random_state=i)
    mod = DecisionTreeClassifier()
    mod.fit(x_train, y_train)
    pred=mod.predict(x_test)
    accu=accuracy_score(y_test,pred)
    if accu>maxAccu:
        maxAccu=accu
        maxRS=i
    print("Best accuracy is ",maxAccu," on Random_state ", maxRS)
```

```python
x_train, x_test, y_train, y_test = train_test_split (x,y,test_size=0.30, random_state=1)
```

# Problem Identification: Since we have to predict the Churn from the given dataset, this is a supervised algorithm. The target variable is Churn column. Since the values are categorical or Y/N, we are applying classification algorithm

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score, accuracy_score, classification_report,confusion_matrix
```

# Fitting DTC Algorithm

```python
DTC = DecisionTreeClassifier()
DTC.fit(x_train, y_train)
pred = DTC.predict(x_test)
print("accuracy train:",DTC.score(x_train,y_train),"\n")
print("accuracy test:",DTC.score(x_test,y_test),"\n")
print("precision :",precision_score(y_test,pred),"\n")
print("Recall :",recall_score(y_test,pred),"\n")
print("f1 score:",f1_score(y_test,pred),"\n")
print("Confusion Matrix \n",confusion_matrix(y_test,pred))
acc=classification_report(y_test,pred)
print(acc)
```

# Fitting RFC Algorithm

```python
RFC = RandomForestClassifier()
RFC.fit(x_train, y_train)
pred = RFC.predict(x_test)
print("accuracy train:",RFC.score(x_train,y_train),"\n")
print("accuracy test:",RFC.score(x_test,y_test),"\n")
print("precision :",precision_score(y_test,pred),"\n")
print("Recall :",recall_score(y_test,pred),"\n")
print("f1 score:",f1_score(y_test,pred),"\n")
print("Confusion Matrix \n",confusion_matrix(y_test,pred))
acc=classification_report(y_test,pred)
print(acc)
```
