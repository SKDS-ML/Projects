```python
from sklearn.preprocessing import OrdinalEncoder
enc=OrdinalEncoder()
```

# Removing Outliers using Zscore

```python
from scipy.stats import zscore

z=np.abs(zscore(df))

new_df=df[(z<3)]
```

# Finding out New Shape

```python
new_df.shape
```

# Finding out Skewness

```python
new_df.skew()
```

```python
x=new_df.drop("label",axis=1)
y=new_df["label"]
```

# Importing Train Test Split

```python
from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score
```

# Identifying Best Random State

```python
for i in range(3):
    x_train, x_test, y_train, y_test = train_test_split (x,y,test_size=0.33, random_state=i)
    from sklearn.tree import DecisionTreeClassifier
    lm=DecisionTreeClassifier()
    lm.fit(x_train,y_train)
    pred=lm.predict(x_test)
    accu=accuracy_score(y_test,pred)
    print(accu*100, "at random_state", i)
```

```python
from sklearn.tree import DecisionTreeClassifier
maxAccu=0
maxRS=0
for i in range(1,3):
    x_train, x_test, y_train, y_test = train_test_split (x,y,test_size=0.30, random_state=i)
    mod = DecisionTreeClassifier()
    mod.fit(x_train, y_train)
    pred=mod.predict(x_test)
    accu=accuracy_score(y_test,pred)
    if accu>maxAccu:
        maxAccu=accu
        maxRS=i
    print("Best accuracy is ",maxAccu," on Random_state ", maxRS)
```

```python
x_train, x_test, y_train, y_test = train_test_split (x,y,test_size=0.30, random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report
```

# Importing Metrics

```python
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score, accuracy_score, classification_report,confusion_matrix
```

# Applying Decision Tree Algorithm

```python
DTC = DecisionTreeClassifier()
DTC.fit(x_train, y_train)
pred = DTC.predict(x_test)
print("accuracy train:",DTC.score(x_train,y_train),"\n")
print("accuracy test:",DTC.score(x_test,y_test),"\n")
print("precision :",precision_score(y_test,pred),"\n")
print("Recall :",recall_score(y_test,pred),"\n")
print("f1 score:",f1_score(y_test,pred),"\n")
print("Confusion Matrix \n",confusion_matrix(y_test,pred))
acc=classification_report(y_test,pred)
print(acc)
```

```python
#Applying Random Forest Algorithm
```

```python
RFC = RandomForestClassifier()
RFC.fit(x_train, y_train)
pred = RFC.predict(x_test)
print("accuracy train:",RFC.score(x_train,y_train),"\n")
print("accuracy test:",RFC.score(x_test,y_test),"\n")
print("precision :",precision_score(y_test,pred),"\n")
print("Recall :",recall_score(y_test,pred),"\n")
print("f1 score:",f1_score(y_test,pred),"\n")
print("Confusion Matrix \n",confusion_matrix(y_test,pred))
acc=classification_report(y_test,pred)
print(acc)
```
