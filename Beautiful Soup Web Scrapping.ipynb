# Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of
release) and make data frame

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
url = "https://www.imdb.com/list/ls091520106/"
page = requests.get(url)
soup = BeautifulSoup(page.content)
name = soup.find_all("h3", class_="lister-item-header")
movies_name = []
for i in name:
    for j in i.find_all("a"):
        movies_name.append(j.text)
year = soup.find_all("span",class_="lister-item-year text-muted unbold")
year_of_release = []
for k in year:
    year_of_release.append(k.text)
rating = soup.find_all("div",class_="ipl-rating-star small")
IMDB_rating = []
for i in rating:
      IMDB_rating.append(float(i.text))
IMDB_top_100=pd.DataFrame({})
IMDB_top_100['movies_name']=movies_name
IMDB_top_100['year_of_release']=year_of_release
IMDB_top_100['IMDB_rating']=IMDB_rating  
IMDB_top_100
```

# Write a python program to scrap book name, author name, genre and book review of any 5 books from
‘www.bookpage.com’

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
page = requests.get('https://bookpage.com/reviews/')
soup = BeautifulSoup(page.content, 'html.parser')
url_tags = soup.find_all("h4",class_="italic")
urls = []
for i in url_tags:
    for j in i.find_all("a", href=True):
        if j.text:
            urls.append(j['href'])
book_name = []
author_name = []
book_genre = [] 
books_review = [] 
for url in urls:
        book = requests.get('https://www.bookpage.com'+url)
        soup = BeautifulSoup(book.content, 'html.parser')
        book_name.append(soup.find('h1').text.replace('\n','').replace('★',''))  
        author_name.append(soup.find('h4').text.replace('\n',''))  
        book_genre.append(soup.find('p', class_="genre-links").text.replace('\n',''))  
        books_review.append(soup.find('div', class_= "article-body").text.replace('\n','')) 
books=pd.DataFrame({})
books["Book Name"]=book_name[:5]
books["Author"]=author_name[:5]
books["Genre"]=book_genre[:5]
books["Review"]=books_review[:5]
books
```

# Top 10 ODI teams in men’s cricket along with the records for matches, points and rating

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
url = "https://www.icc-cricket.com/rankings/mens/team-rankings/odi"
page = requests.get(url)
soup = BeautifulSoup(page.content)
team = soup.find_all("span",class_='u-hide-phablet')
team_name = []
for i in team:
    team_name.append(i.text)
matches = [] 
points = [] 
ratings = [] 
new_list = [] 
for i in soup.find_all("td",class_='rankings-block__banner--matches'): 
    matches.append(i.text)
for i in soup.find_all("td",class_='rankings-block__banner--points'):
    points.append(i.text)
for i in soup.find_all("td",class_='rankings-block__banner--rating u-text-right'):
    ratings.append(i.text.replace("\n",""))
for i in soup.find_all("td",class_='table-body__cell u-center-text'):
    new_list.append(i.text)
for i in range(0,len(new_list)-1,2):
    matches.append(new_list[i]) 
    points.append(new_list[i+1]) 
for i in soup.find_all("td",class_='table-body__cell u-text-right rating'):
    ratings.append(i.text)
icc=pd.DataFrame({})
icc['Team_name']=team_name[:10]
icc['Matches']=matches[:10]
icc['Points']=points[:10]
icc['Ratings']=ratings[:10]
icc    
```

# Write a python program to extract information about the local weather from the National Weather Service
website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
url = "https://forecast.weather.gov/MapClick.php?textField1=37.78&textField2=-122.42#.YJJVsLUzZPY"
page = requests.get(url)
soup = BeautifulSoup(page.content)
period = []  
short_desc = []   
temperature = []  
description = []    
for i in soup.find_all("div", class_="col-sm-2 forecast-label"):
    period.append(i.text)
for i in soup.find_all('div',class_="col-sm-10 forecast-text"):        
        short_desc.append(i.text.split(',')[0])
for i in soup.find_all('div',class_="col-sm-10 forecast-text"):        
        description.append(i.text)
for i in soup.find_all('p',attrs={'short-desc'}):
        if i.next_sibling is not None:
            temperature.append(i.next_sibling.text)
        else:
            temperature.append(' ')
for i in soup.find_all("div", class_="col-sm-10 forecast-text"):
        description.append(i.text)
san_francisco_weather=pd.DataFrame({})
san_francisco_weather['Period']=period[:9]
san_francisco_weather['Temperature']=temperature[:9]
san_francisco_weather['Short_description']=short_desc[:9]
san_francisco_weather['Description']=description[:9]
san_francisco_weather
```

# Write a python program to scrape fresher job listings from ‘https://internshala.com/’. It should include job title,
company name, CTC, and apply date

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
url = "https://internshala.com/fresher-jobs"
page = requests.get(url)
soup = BeautifulSoup(page.content)
job_titles = []  
company_names = []  
CTC = []   
apply_date = []  
titles=soup.find_all('div',class_="heading_4_5 profile") 
for i in titles:
    job_titles.append(i.text.replace('\n', ''))  
companies=soup.find_all('a',class_="link_display_like_text")
for i in companies:
    company_names.append(i.text.strip())
job_info = []
info = soup.find_all("div",class_="item_body")
for i in info:
    job_info.append(i.text.replace("\n","").replace("\xa0",""))   
for i in range(2,len(job_info),3):
    apply_date.append(job_info[i].strip())
for i in range(1,len(job_info),3):
    CTC.append(job_info[i].strip())
JOB=pd.DataFrame({})
JOB['Job_Titles']=job_titles
JOB['Company_Names']=company_names
JOB['CTC']=CTC
JOB['Apply_Date']=apply_date
JOB
```
